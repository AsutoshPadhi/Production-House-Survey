```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(10)
```

```python
(pd.crosstab(df['Gender'], df['Frequency'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown'],grid=False)
plt.show()
```

```python
(pd.crosstab(df['Occupation'], df['Frequency'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown'],grid=False)
plt.show()
```

```python
(pd.crosstab(df['Gender'], df['Language'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown','yellow','lightblue','coral','crimson','cyan'],grid=False)
plt.show()
```

This shows that males prefer pure English more than any other language wheras females prefer a mixture of English, Marathi and Hindi. 
Also, only men watch Japanese

```python
(pd.crosstab(df['Gender'], df['Genre'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown','yellow','lightblue','coral','crimson','cyan','fuchsia'],grid=False)
plt.show()
```

Here we can see that Males in general prefer comedy over every other genre wheras females dont have a specific preference as such. 
Also it can be seen that some genres like Romance and Horror is generally seen by females so this should be targetted for female audience.
There is a general conception that females fancy family drama which is not true in this case. This age group prefers otherwise.

```python
(pd.crosstab(df['Gender'], df['Type'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown','yellow','lightblue','coral','crimson','cyan','lavender','lime','magenta','maroon','navy','orange'],grid=False).legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.show()
```

```python
(pd.crosstab(df['Gender'], df['Channels'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown','yellow','lightblue','coral','crimson','cyan','lavender','lime','magenta','maroon','navy','orange'],grid=False).legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.show()
```

It is very clear from this that males watch a lot of Zee Cafe, Romedy Now and Comedy Central.

```python
(pd.crosstab(df['Gender'], df['Time'])).plot(kind='bar',stacked=False,color=['red','blue','green','brown','yellow','lightblue','coral','crimson','cyan','lavender','lime','magenta','maroon','navy','orange'],grid=False).legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.show()
```

From the earlier graphs we have seen that males generally prefer comdey over every other genre. So Channels should put on comedy shows from 9pm to 12am since that is when males often watch TV.
We had also seen that females prefer a combination of Comedy and Romance. Since females generally tend to watch TV from 6pm - 9pm, that is when Channels should put on Comedy and Romance.

It was also seen that a lot of males prefer Quiz shows wheras a lot of females prefer Talent Shows. Thus the target audience for the 9pm-12am slot wants to see a mixture of Comedy and Quiz. And the target audience of 6pm-9pm wants to see a mixture of comdey, horror and romance in Talent shows, Quiz shows and Chit-chat shows.
